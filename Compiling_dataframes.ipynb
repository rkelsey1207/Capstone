```python
tweets_2015= pd.read_csv('./final_tweets_2015')
tweets_2016 = pd.read_csv('./final_tweets_2016')
tweets_2017 = pd.read_csv('./final_tweets_2017')

final_2015 = pd.read_csv('./datasets/game_2015')
final_2016 = pd.read_csv('./datasets/game_2016')
final_2017 = pd.read_csv('./datasets/game_2017')

accounts_2017 = pd.read_csv('./datasets/twitter_accounts.csv')
accounts_2016 = pd.read_csv('./datasets/account_2016.csv')
accounts_2015 = pd.read_csv('./datasets/account_2015.csv')
```

```python
tweets_2015['date'] =  pd.to_datetime(tweets_2015['date'],format= "%Y/%m/%d")
tweets_2016['date'] =  pd.to_datetime(tweets_2016['date'],format= "%Y/%m/%d")
tweets_2017['date'] =  pd.to_datetime(tweets_2017['date'],format= "%Y/%m/%d")

final_2015['date'] =  pd.to_datetime(final_2015['date'],format= "%Y/%m/%d")
final_2016['date'] =  pd.to_datetime(final_2016['date'],format= "%Y/%m/%d")
final_2017['date'] =  pd.to_datetime(final_2017['date'],format= "%Y/%m/%d")
```

```python
tweets_2015.drop(['Unnamed: 0'], axis=1, inplace=True)
tweets_2016.drop(['Unnamed: 0'], axis=1, inplace=True)
tweets_2017.drop(['Unnamed: 0'], axis=1, inplace=True)

```

```python
def get_week(tweet_df): 
    nfl_week = []
    for i in tweet_df['date']:
        yr, weekof, dayof = i.isocalendar()
        week = (weekof - 1) if dayof == 1 else weekof
        nfl_week.append(week)
    tweet_df['nfl_week']= nfl_week
    return
get_week(tweets_2015)
get_week(tweets_2016)
get_week(tweets_2017)
get_week(final_2015)
get_week(final_2016)
get_week(final_2017)
```

```python
def tweet_info(df, final):
    compound = df.groupby(['nfl_week','name'], as_index=False)['compound'].mean() 
    weekly_tweets = df.groupby(['nfl_week','name'], as_index=False)['text'].count()
    neg_tweets = df.groupby(['nfl_week','name'], as_index=False)['neg'].agg('max') 
    pos_tweets = df.groupby(['nfl_week','name'], as_index=False)['pos'].agg('max') 
    stdev = df.groupby(['nfl_week','name'], as_index=False)['compound'].agg(np.std, ddof=0)
    return
tweet_info(tweets_2015, final_2015)
```

```python
compound = tweets_2015.groupby(['nfl_week','name'], as_index=False)['compound'].mean() 
weekly_tweets = tweets_2015.groupby(['nfl_week','name'], as_index=False)['text'].count()
neg_tweets = tweets_2015.groupby(['nfl_week','name'], as_index=False)['neg'].agg('max') 
pos_tweets = tweets_2015.groupby(['nfl_week','name'], as_index=False)['pos'].agg('max') 
stdev = tweets_2015.groupby(['nfl_week','name'], as_index=False)['compound'].agg(np.std, ddof=0)
neg_avg = tweets_2015.groupby(['nfl_week','name'], as_index=False)['neg'].mean() 
pos_avg = tweets_2015.groupby(['nfl_week','name'], as_index=False)['pos'].mean()
neg_avg.rename(columns={'neg':'neg_avg'}, inplace=True)
pos_avg.rename(columns={'pos':'pos_avg'}, inplace=True)
```
