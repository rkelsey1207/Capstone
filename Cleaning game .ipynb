## Cleaning Fantasy DF

```python
def clean_fantasy(df):
    df = df[df['Rk'] <= 150]
    df['Player'] = df['Player'].str.rstrip('*')
    df['Player'] = df['Player'].str.rstrip('+')
    df['ppr_avg']= df['PPR'] / df['G']
    df['FantPt_avg']= df['FantPt'] / df['G']
    return 
clean_fantasy(fantasy_2015)
```
